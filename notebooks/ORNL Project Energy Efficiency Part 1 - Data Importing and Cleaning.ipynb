### Missing Data

Missing data is a common problem when it comes to real data sets. There can be multiple reasons for their occurences but some of the main ones are:

- Human Error
- Equipment fault
- Unexpected event
- Data loss through processing

The following function takes in a dataframe and produces a table with the absolute number and percentage of missing values. This function was adapted from [this post](https://stackoverflow.com/questions/26266362/how-to-count-the-nan-values-in-a-column-in-pandas-dataframe/39734251#39734251).

#### Missing Data Table

```python
# Function to calculate missing values by column
def missing_values_table(df):
    """
    Takes a dataframe as input and returns
    a table with percentage of missing values
    and absolute number per column
    
    """
    
    # Calculate all the missing values
    mis_val = df.isnull().sum()
    
    # Percentage of missing values
    mis_val_percent = 100 * df.isnull().sum() / len(df)

    # Make a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    # Rename the columns
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})

    # Sort the table by percentage of missing descending
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
    '% of Total Values', ascending=False).round(1)

    # Print some summary information
    print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
          " columns that have missing values.")

    # Return the dataframe with missing information
    return mis_val_table_ren_columns
```

```python
# Produce the table
missing_df = missing_values_table(df_main);

display(missing_df)
```

Getting rid of data is always dangerous as you might be letting go useful information however in this case there is too much data missing to be able to get any use out of those columns. We will therefore proceed to get rid of any columns that have > 50% missing values. 

The following function, takes care of this and also reports the definitions of the columns that were drop to get an idea of what kind of data we are losing. 

```python
def remove_missing_values(threshold, missing_value_table, df):
    """
    Removes the columns of your data based on 
    a given threshold of missing percentage
    
    """
    
    # Get columns above threshold
    missing_columns = list(missing_df[missing_df['% of Total Values'] > threshold].index)
    print('We will remove %d columns.' % len(missing_columns))
    
    # Definitions of the missing columns
    print()
    define_columns(missing_columns)

    # Drop the columns
    df = df.drop(columns = list(missing_columns))
    
    return df
```

#### Filling the Missing Data

In this section we will use the columns from the above missing data table to understand the cause of missing information and find an appropriate way to fill it in. We will first look at the four columns that only have 1 value missing. 

##### Features with negligible missing values
The following script takes in the missing table, finds the columns with less that 0.1% percent missing data and replaces them by interpolation. The reason we are using interpolation is to maintain the natural "continuity" of time series data. This basically assumes that within the 15 minute interval of missing data there wasn't much change in the measurements which is a fair assumption when considering the kind of measurements we are taking.

```python
def fill_negligible(missing_table, df, threshold = 0.1):
    """
    Takes in a missing table and finds the columns
    with less that threshold of missing percentage
    It then replaces them by interpolation
    """
    
    # Get the features that only have 1 missing value 
    negligible_missing_columns = list(missing_table[missing_table["% of Total Values"] < 0.1].index)
    
    # Go through each column and replace NaN values with interpolation
    for col in negligible_missing_columns:
        
        # Interpolate
        df[col] = df[col].interpolate()
        print("Replaced NaN values of column : {}".format(col))
        
    return df
```

The following function goes through all the columns we are interested in and finds which indexes have missing values and which not by creating a binary columns of 1 indicating that the index is missing some data and 0 it doesn't. It then builds a new dataframe with just the binary columns and plots it. The resulting visualization allows us to actually see the spread of missing values along our entire dataset. 

```python
# Define the function to plot the spread of missing values
#---------------------------------------------------------

def plot_missing_spread(df, columns):

    """
    Takes in a dataframe and the columns we are
    interested to find out the spread of their missing
    values. It then plots a subplot that represents the
    spread of each missing value along the entire dataset

    """
    # Define the dictionary to be populated
    d = {}

    # Loop through all columns
    for col in columns:
        
        # Get the binary column of missing values and add it to the dictionary
        d[col]= df[col].isna().astype(int)

    # Create a new df to plot with all binaries
    plot_df = pd.DataFrame(d)
    
    if plot_df.shape[1] == 1:
        
        # Plot a barplot
        ax = plot_df.plot(subplots=True, layout=(1, 1), figsize=(10, 6), sharex=True, title = "Missing Values Spread");
    else:
        # Plot a subplot
        ax = plot_df.plot(subplots=True, layout=(int(plot_df.shape[1]/2), 2),
                          figsize=(16, 12), sharex=True, title = "Missing Values Spread")
```

So we need to define a function that carries out some checks.  

We will now define a function that gets all the index numbers of the missing values and determines the difference between them and calculates some important statistics. It also calculates the number of consecutive values of missing data by finding taking into consideration values that have index difference of 1 between them.

```python
from collections import defaultdict

def missing_data_statistics(df,columns):
    
    """
    
    Takes in a dataframe and the columns with
    the missing values and finds some basic statistics
    such as min, max, std and percentiles for the 
    
    """
    
    # Create a copy of the dataset
    df_copy = df.copy()
    
    d = defaultdict(list)
    
    # Loop through each column
    for col in columns:
        
        # Get the index of missing values
        index = df_copy[col].index[df_copy[col].apply(np.isnan)]
        
        index_series = pd.Series(index).diff()
        
        # consecutive counter
        cons = []
        
        
        d["Stats"] = ["Consecutive Count", "Min", "Max", "Std", "Median", "Mean"]
        
        for i, ind in enumerate(list(index)):

            if i == 0:
                continue
            
            # Find the difference
            diff = list(index)[i] - list(index)[i - 1]
            
            # Increment number of consecutives
            if diff == 1:
                cons.append(ind)
            
        # Calculate statistics and add to dictionary
        d[col].append(int(len(cons)))
        d[col].append(round(index_series.min(),2))
        d[col].append(index_series.max())
        d[col].append(index_series.std())
        d[col].append(index_series.median())
        d[col].append(index_series.mean())

    stats_df = pd.DataFrame(d)   
    stats_df = stats_df.set_index("Stats")
    stats_df = stats_df.round(2)
    
    return stats_df
```

The larger mean than the median suggest that the data is left skewed, or that there are some very large values in the dataset that interfere with the mean. This is okay though as it only means that some missing occurrences have very large differences between them, i.e. random. From the results of the above as we only have 11 consecutive data points and we also have a median of 2 values space between missing values, it is safe to assume that overall the missing data is due to less frequent measurements and non-long term faults so we can use interpolation.

```python
def get_time_period(time_column):
    
    # Get the first and last date
    start_dt = time_column.min()
    end_dt = time_column.max()
    
    # Calculate the differences in various units
    day_diff = round((end_dt - start_dt)/np.timedelta64(1,'D'),2)
    month_diff = round((end_dt - start_dt)/np.timedelta64(1,'M'),2)
    year_diff = round((end_dt - start_dt)/np.timedelta64(1,'Y'),2)
    
    # Report the period span
    print(f"The first date is : {start_dt}")
    print(f"The last date is : {end_dt}")
    print()
    print("The data spans across {} Years / {} Months / {} Days".format(year_diff, month_diff, day_diff))
```

```python
# Get the period
get_time_period(df_cleaned["TIMESTAMP"])
```

As we can see we have data for 1 year in total from October 2013 to October 2014.
