### Converting Cyclic Features

Cyclic features as the name suggests are mostly common with time series data, they are features that have a "cyclic" i.e. they turn to reach a number / state and then repeat themselves. The best example for this is the hours in a day. The model has no way to "understand" that 23 comes after 0 as it looks at everything mathematically. To preserve this cyclic behavior we will convert these features into cyclic features by using cos and sin variations. This method is described in detail in [this article.](http://blog.davidkaleko.com/feature-engineering-cyclical-features.html)


![circle](img/circle.png)

The following function will take any cyclic variable and given a time period it will map it onto a circle. What we mean by that, it will represent the same variable as an x and y coordinate of a circle. This allows the continuity of cyclic features to be picked up by the model. We need to calculate two new features of each cyclic variable, its cosine (x) and sin(y). The diagram above illustrates this. 

```python
# We first need to get the numeric representation
# of our cyclic values
df_processed["Month"] = df_processed.index.month
df_processed["Day"] = df_processed.index.dayofweek

display(df_processed[["Day", "Month"]].head())


# df_processed.index.month.min()
```

Good we now need to define all columns and their periods. Note that for all columns that start from 1 instead of 0 we subtract one for simplicity

```python
# Define a diction with the cyclic features 
# of our dataset and their period
cyclic_d = {
    "Month" : 12,
    "Day" : 7,
    "Week" : 52,
    "Hour" : 24
}

# Define the function to convert cyclic values
def convert_to_cyclic(df, col, period):
    
    """
    Takes in a column of cyclic
    nature and converts the values
    to cyclic by taking the sin
    and cosine. The period is required
    
    """
    
    # Check if lowest value isn't zero
    if df[col].unique().min() > 0:

        # Substract one
        df[col] = df[col] - 1
    
    # Create the new values
    df[f"{col}_sin"] = np.sin(df[col] * (2. * np.pi/period))
    df[f"{col}_cos"] = np.cos(df[col] * (2. * np.pi/period))
    
    return df
```

```python
# Loop through all cyclic values
for val, period in cyclic_d.items():
    
    df_processed = convert_to_cyclic(df_processed, val, period)
    
display(df_processed.head())
```

### Splitting the Dataset

The last thing to do before we can start modeling is split the dataset into training and test variables. The following function taken from [here](https://nbviewer.jupyter.org/github/Yorko/mlcourse_open/blob/master/jupyter_english/topic09_time_series/topic9_part1_time_series_python.ipynb) does just that. We just need to provide our X values and y values and the test ratio we want. 

```python
def timeseries_train_test_split(X, y, test_size):
    """
        Perform train-test split with respect to time series structure
    """
    
    # get the index after which test set starts
    test_index = int(len(X)*(1-test_size))
    
    X_train = X.iloc[:test_index]
    y_train = y.iloc[:test_index]
    X_test = X.iloc[test_index:]
    y_test = y.iloc[test_index:]
    
    return X_train, X_test, y_train, y_test
```

```python
# Split the dataset with 20% reserved for testing 
X_train, X_test, y_train, y_test = timeseries_train_test_split(scaled_df, y, test_size = 0.2)
```

## Modeling

Before we proceed with testing different models we first need to establish a baseline. With a regression problem a good baseline is to "assume" all values are equal to the median. We also need to define our metric which will be MAE. 

### Defining a metric

Mean absolute error, is a good metric to use for our problem as we can directly related the answer to the problem statement. What we mean by that, is that the error is a direct measurement of the error in energy usage that we should expect when using the model to predict. 

```python
# Function to calculate mean absolute error
def mae(y_true, y_pred):
    """
    Calculates MAE between
    y_true values and predictions
    """
    return np.mean(abs(y_true - y_pred))

def mean_absolute_percentage_error(y_true, y_pred): 
    
    """
    This function calculates the MAE 
    percentage error
    
    """
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

### Establishing a baseline

We will create a "baseline" result that we need to "compete" against. This is a good way to understand how well our model is performing. 

```python
# Obtain the median of the Target value as a guess
baseline_guess = np.median(y)

print('The baseline guess is a score of %0.2f' % baseline_guess)
print("Baseline Performance on the test set: MAE = %0.4f" % mae(y_test, baseline_guess))
```

The above result indicates that if we just "predict" that the energy will be the Median of all measurements up to now we will have an error of +/- 123.9 kWh when predicting the energy usage. This is the value that we will attempt to beat. 

### Testing various algorithms

In this part of the model we will define a function that goes through the following four algorithms, calculates the score and plots the results. 

- Linear Regression
- Random Forest
- KNearest Neighbor
- Gradient Boost

The reason we chose these four algorithms is because they have are quite easy to get so interpretation out of the results. Where with models such as SVR this is always the case. In this study it is quite important to understand the working behind our algorithm and what features drove it to its prediction. 

#### Defining the Functions

Here we will define two functions that will loop through the four algorithms mentioned and plot the results to get a direct comparison. 

```python
# Define function to evaluate a model
def fit_and_evaluate(model, X_train, X_test, y_train, y_test):
    
    """
    Takes in all the parts of a dataset and a specific
    model defined and returns the MAE of that model
    
    """
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions and evalute
    model_pred = model.predict(X_test)
    model_mae = mae(y_test, model_pred)
    
    # Return the performance metric
    return model_mae
```

Function that runs the four models

```python
from collections import OrderedDict

def run_models(X_train, X_test, y_train, y_test):
    
    """
    This function takes in the splits of the 
    dataset, defines the four models below and
    fits the data to them. It then plots the 
    results of the MAE.
    
    Models :
    
    - LR
    - RF
    - GB
    - KNN
    
    """
    
    # Define the models list
    models_list = []
    
    # Define the MAE dictionary
    models_d = defaultdict(list)
    
    # Add the model names
    models_names = ["Linear Regression", "Gradient Boost", "Random Forest", "KNN"]
    
    models_d["Model"] = models_names
    
    # Define the models with basic hyperparameters
    lr = LinearRegression()
    gbr = GradientBoostingRegressor()
    rf = RandomForestRegressor()
    knn = KNeighborsRegressor(n_neighbors=10)
    
    # Add the models to the list
    models_list.append(lr)
    models_list.append(gbr)
    models_list.append(rf)
    models_list.append(knn)
    
    for name,model in zip(models_names, models_list):
        
        # Get the MAE
        model_mae = fit_and_evaluate(model, X_train, X_test, y_train, y_test)
        
        # Add the MAE to the dictionary
        models_d["MAE"].append(model_mae)
        
        # Print progress
        print(f"Finished running {name} and the MAE is : {model_mae}")
    
    
    # BUILDING THE COMPARISON
    #------------------------
    
    # Create a dataframe with the results
    model_comparison = pd.DataFrame(models_d)
    
    # Plot a horizontal plot that compares all the MAEs
    model_comparison.sort_values('MAE', ascending = False).plot(x = 'Model', y = 'MAE', kind = 'barh',
                                                           color = 'orange', edgecolor = 'black',
                                                                alpha = 0.8, figsize = (10,5))
    
    
    # Plot formatting
    plt.ylabel(''); plt.yticks(size = 14); plt.xlabel('Mean Absolute Error'); plt.xticks(size = 14)
    plt.title('Model Comparison on Test MAE', size = 20);
    plt.axvline(x= mae(y_test, baseline_guess));
```

From the above it seems that KNN the best model to use. However, KNN doesn't have as many parameters to tune therefore, this result is probably as good as it will get. We will therefore go with the second best, which is the Gradient Boost and try to optimize the result.

### Optimizing  the Algorithm

In this section we will use the picked model from before, Gradient Boost and run a hyper-parameter tuning technique. We will use a K-Fold validation technique specific used for timeseries that splits the data in K different parts. We will then define some hyper parameters that can be modified for this algorithm and run a RandomizedSearchCV. This will run a certain number of iterations using various variations of the parameters and it will then report the best combination. 

Most of the methods used here were taken fro [this article](https://github.com/WillKoehrsen/machine-learning-project-walkthrough/blob/master/Machine%20Learning%20Project%20Part%202.ipynb) as it summarizes these steps very well. We have adapted this for our case.

```python
# Set the TimeSeries split into 5 splits
tscv = TimeSeriesSplit(n_splits=5)
```

#### Running a RandomSearchCV 

In this section we will define the different parameters to check for the RandomForestClassifier algorithm. The parameters were chosen after looking in the scikit learn documentation.

```python
# Number of trees used in the boosting process
n_estimators = [100, 200, 300, 400]

# Maximum depth of each tree
max_depth = [2, 3, 5, 10, 15]

# Minimum number of samples per leaf
min_samples_leaf = [1, 2, 4, 6, 8]

# Minimum number of samples to split a node
min_samples_split = [2, 4, 6, 10]

# Maximum number of features to consider for making splits
max_features = ['auto', 'sqrt', 'log2', None]

# Define the grid of hyperparameters to search
hyperparameter_grid = {'n_estimators': n_estimators,
                       'max_depth': max_depth,
                       'min_samples_leaf': min_samples_leaf,
                       'min_samples_split': min_samples_split,
                       'max_features': max_features}
```

We now need to define the model that will run the K-Fold validation and define the random search. The following script does just that. We have made the following adjustments to the random search:

- We will run 25 different model combinations
- We will use the time series split
- The model we are using is Gradient Boost

```python
# Create the model to use for hyperparameter tuning
model = RandomForestRegressor()

# The random search below will run 
random_cv = RandomizedSearchCV(estimator=model,
                               param_distributions=hyperparameter_grid,
                               cv=tscv, n_iter=25, 
                               scoring = 'neg_mean_absolute_error',
                               n_jobs = -1, verbose = 10, 
                               return_train_score = True)
```

#### Running a GridSearchCV on N_estimators

Ideally to get the absolute best model we need to run a GridSearch rather than a RandomSearch. The difference is that a grid search runs every possible hyperparameter that we specified. The large number of possibilities requires a significant amount of processing power / runtime. For now we will run a GridSearchCv on just the n_estimators / number of trees hyper parameter. 

The effect of using more trees can usually lead to a better result as the more trees used the more "possibilities", however a common problem with this is that as the number of trees increase as some point we start seeing overfitting kicking in. This is when the model fits our data "too well" to be able to generalise. We will try to visualise this trade-off by plotting the training error and test error. 

The following script runs a GridSearch on various number of trees / estimators.

##### Defining and Fitting the GridSearchCV

```python
# Define the number of trees to test
trees_grid = {'n_estimators': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800]}

model = RandomForestRegressor(    max_depth = 15,
                                  min_samples_leaf = 4,
                                  min_samples_split = 6,
                                  max_features = "auto")

# Grid Search Object using the trees range and the random forest model
grid_search = GridSearchCV(estimator = model, param_grid=trees_grid, cv = tscv, 
                           scoring = 'neg_mean_absolute_error', verbose = 10,
                           n_jobs = -1, return_train_score = True)
```

#### Model Results Function

The following function presents the results of a defined model using the training and test datasets. The function carries out the following actions:

- Calculates the MAE and percentage MAE of the final model
- Plots a KDE plot of prediction and actual data
- Plots the residuals of the model

```python
rf = grid_search.best_estimator_

def regression_results(model, X_train, X_test, y_train, y_test):

    
    # GET THE PREDICTIONS
    #--------------------
    
    # Get the baseline guess
    baseline_guess = np.median(y_train)
    mae_baseline = mae(y_test, baseline_guess)
    
    # Obtain the predictions of the model
    final_pred_train = model.predict(X_train)
    final_pred_test = model.predict(X_test)

    # Calculate MAE and Mean Absolute Percentage Error for train and test set
    mae_final_train = mae(y_train, final_pred_train)
    mae_perc_train = mean_absolute_percentage_error(y_train, final_pred_train)
    
    mae_final_test = mae(y_test, final_pred_test)
    mae_perc_test = mean_absolute_percentage_error(y_test, final_pred_test)
    
    
    # Print the MAE and the Mean Absolute Percentage Error
    print('The baseline guess is a score of %0.2f' % baseline_guess)
    print("Baseline Performance on the test set: MAE = %0.4f" % mae_baseline)
    print()
    print("Final model performance on the train set:   MAE = {0:.3f}".format(mae_final_train))
    print("Final model performance on the train set:   Mean Absolute Percentage Error = {0:.1f}%".format(mae_perc_train))
    print()
    print("Final model performance on the test set:   MAE = {0:.3f}".format(mae_final_test))
    print("Final model performance on the test set:   Mean Absolute Percentage Error = {0:.1f}%".format(mae_perc_test))


    
    print()
    print()
    print("---------------------------------------------------")
    print("Plotting the KDE Results and Residuals of the Model")
    print("---------------------------------------------------")
    print()
    
    
    # PLOT THE KDE PLOTS
    #-------------------
    
    
    # Define a plot size
    plt.figure(figsize = (14,6))
    
    # Density plot of the final predictions and the test values
    sns.kdeplot(final_pred, label = 'Predictions')
    sns.kdeplot(y_test, label = 'Truth')

    # Label the plot
    plt.xlabel('Mains Energy (kWh)'); plt.ylabel('Density');
    plt.title('Test and Truth Density Plots Comparison (Test Set)');
    
    # PLOT THE RESIDUAL PLOTS
    #------------------------
    
    # Define the figsize
    plt.figure(figsize = (14,6))
    
    # Calculate the residuals of the predictions
    residuals = final_pred - y_test
    
    # Plot the residuals in a histogram
    plt.hist(residuals, color = 'red', bins = 50,
             edgecolor = 'black')
    
    plt.xlabel('Error'); plt.ylabel('Count')
    plt.title('Distribution of Residuals for Final Model (Test Set)');
```

```python
def plot_interactive_twinaxis(df, f1, f2):
    
    """
    
    Takes in a dataframe and plots both of them
    at the same axis
    
    f1 is the feature to be plotted on the
    first y-axis. 
    
    f2 is the feature to be plotted on 
    the same axis
    
    """
   
    # BUILDING THE DATA
    #------------------
    
    # data on primary y-axis
    y1_trace = go.Scatter(

        x = df.index,
        y = df[f1],
        name = f1,
        line = dict(color = 'rgb(117, 155, 193)'),
        opacity = 0.9

    )      
    
    # Data on secondary y-axis
    y2_trace = go.Scatter(

        x = df.index,
        y = df[f2],
        name = f2,
        line = dict(color = 'rgb(193, 66, 66)'),
        opacity = 0.7
    )

    # Add the two data traces
    data = [y1_trace, y2_trace]

    
    # PLOT LAYOUT
    #------------
    
    layout = go.Layout(
        
        title = f'{f1} Plot by {f2}',
        
        yaxis = dict(
            
            title = f'{f1}'
            
            ),
        
        xaxis=dict(
            
        rangeselector=dict(
            
            buttons=list([
        
                dict(count = 12,
                     label ='12h',
                     step = 'hour',
                     stepmode = 'backward'),
                
                dict(count = 1,
                     label = '1d',
                     step='day',
                     stepmode='backward'),
                
                dict(count=7,
                     label='1w',
                     step='day',
                     stepmode='backward'),
                
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                
                dict(step='all')
                
                ])
            ),
            
        rangeslider=dict(
            
            visible = True
        ),
        type='date'
        )

    )

    fig = go.Figure(data=data, layout=layout)
    iplot(fig, filename='multiple-axes-double')
```

### Model Evaluation / Interpretation

The way that Lime works is by learning a linear regression model around the current model which is easy to interpret. 

#### Feature Importance

The first thing we will look at is feature importance. This will tell us which feature "contributes" most to the prediction of the model. This not only tell us which features are the key ones that let us predict energy usage but also which one aren't as important and perhaps could have reached the same result without them. This can save us processing power as the more features the more time it takes to run a model. Luckily, ensemble models such as Random Forest , allow us to explore this quite easily. 

```python
# Define function for importances
#---------------------------------

def get_feat_importance(model, X_train, plot_feat = False, max_feat = 10):
    
    # Create a df with feature importances
    feature_df = pd.DataFrame({"Features" : list(X_train.columns),
                               "Importance" : model.feature_importances_})
    
    # Sort the results by top performers
    feature_df = feature_df.sort_values('Importance', ascending = False).reset_index(drop=True)

    print(f"Top {max_feat} Features for this Model")
    # Display the Top 10
    display(feature_df.head(max_feat))
    
    # If plot_feat is true then plot the results
    if plot_feat:
        
        print()
        print("--------------")
        print("Features Plot")
        print("--------------")
        # Create the plot
        feature_df.loc[:max_feat - 1, :].plot(x = 'Features', y = 'Importance', 
                                     edgecolor = 'k', alpha = 0.7,
                                     kind='barh', color = 'green', figsize = (14,6));
        
        # Plot characteristics
        plt.xlabel('Relative Importance', size = 15); plt.ylabel('')
        plt.title(f'Feature Importances for Top {max_feat} Features', size = 20);
        
    return feature_df
```

```python
feature_df = get_feat_importance(model = final_model, X_train = X_train, plot_feat = True, max_feat = 15)
```

#### Explaining the Model

The final section will look at two prediction examples, one that the model performed poorly and one that did well. We will then try to understand what caused this performance change by using LIME. The following function takes in a model and the most important features and builds a LIME explainer. It then detects two predictions, a good one and bad one and explains the performance by plotting how the model "came to" the final result.

```python
def explain_model_performance(model, predictions, y_test, X_test, X_train):
    
    # DETECT A GOOD AND A BAD PREDICTION
    #-----------------------------------
    
    # Calculate the residuals of the model
    residuals = predictions - y_test
    
    # Get the best and worst prediction
    worst_pred = X_test.loc[residuals.idxmax, :]
    best_pred = X_test.loc[np.argmin(np.abs(residuals)), :]
    
    
    # CREATE AN EXPLAINER MODEL
    #--------------------------
    
    # Build an explainer model that uses the same data as the original model
    explainer = lime.lime_tabular.LimeTabularExplainer(training_data = X_train.values, 
                                                   mode = 'regression',
                                                   training_labels = y,
                                                    feature_names = list(X_train.columns))
    
    
    # EXPLAIN WORST PREDICTION
    
    # Display the predicted and true value for the wrong instance
    print("Worst Prediction Information")
    print("----------------------------")
    print()
    print('Prediction: %0.4f' % model.predict(worst_pred.values.reshape(1, -1)))
    print('Actual Value: %0.4f' % y_test[residuals.idxmax])
    print()

    # Run the explainer on the worst prediction
    wrong_exp = explainer.explain_instance(data_row = worst_pred, 
                                           predict_fn = model.predict)

    # Plot the prediction explaination
    wrong_exp.as_pyplot_figure(figsize(10,7));
    plt.title('Explanation of Worst Prediction', size = 18);
    plt.xlabel('Effect on Prediction', size = 14);
    
    # EXPLAIN WORST PREDICTION
    
    # Display the predicted and true value for the wrong instance
    print("Best Prediction Information")
    print("----------------------------")
    print()
    # Display the predicted and true value for the best prediction
    print('Prediction: %0.4f' % model.predict(best_pred.values.reshape(1, -1)))
    print('Actual Value: %0.4f' % y_test[np.argmin(np.abs(residuals))])
    print()

    

    # Explanation for the best prediction
    right_exp = explainer.explain_instance(best_pred, model.predict, num_features=10)
    
    
    # Plot the best prediction explanation plot
    right_exp.as_pyplot_figure(figsize(10,7));
    plt.title('Explanation of Best Prediction', size = 18);
    plt.xlabel('Effect on Prediction', size = 14);
```

```python
explain_model_performance(final_model, final_pred, y_test, X_test, X_train)
```

**Comments**

The above plots simply show what feature contributions made the model make the choice it did :

- Starting with the worst prediction, first of all we can see that the model over predicted the energy usage by almost four times the true value. The graph below shows us what let the model to decide that the energy should be so high. The value of the previous measurement, four measurements and eight measurements ago was more than 540 which let the model think that the current measurement should be quite high as well. This makes sense if we try to think of a model as a human person that looks at what happened before and just takes an "educated" guess. What probably happened in this case, there must have been a sudden drop in energy usage that the model couldn't have predicted, perhaps an HVAC system was switched off. 


- The second case, shows an example where the model predicted the energy usage almost exactly right. The logic is quite similar though, the measurements mentioned before were quite low (<= 257) therefore the model went for the logical value of 171. 

This shows us that the model can predict quite well in cases where the energy usage varies smoothly, however in cases where there are sudden drops or increases the model won't do as well. Perhaps if we had metadata on HVAC schedule we could have added that as a new feature and the model would have taken that into account. I.e. if the heating was scheduled to turn on at 17:00 the model would expect a spike around 17:00.

3.
